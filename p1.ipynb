```python
import pandas as pd
```

```python
df = pd.read_csv("poke.csv")
```

```python
df.head()
```

```python
# to count values in file -> rows & cols

# df.count()
# df.info()
df.shape
```

```python
# name of max speed
df.loc[df["Speed"]==45]["Name"]
```

```python
# hp where name = Bulbasaur
df.loc[
    df["Name"]=="Bulbasaur"
]["HP"]
```

```python
# max & min value by sorting

# df.sort_values(by=["Speed"],ascending=False)
df.sort_values(by=["Speed"])
```

```python
# Speed of most common name

# df["Name"].describe()
df["Name"].mode()
# df.loc[df["Name"]==df["Name"].mode()]["Speed"].values[0]
# df.loc[df["Name"]==df["Name"].mode()]["Speed"]
df['Speed'].value_counts().head()
```

```python
# slowest speed of hp>50 & name also
# df.loc[df['HP']>50].sort_values(by='Speed').head(1)
df.loc[df['HP']>50].sort_values(by='Speed').head(1)['Name']
```

```python
# setting a column as index

df.set_index('Num')
```

```python
# create col ratio which is ratio of attack / speed

df['ratio'] = df['Attack'] / df['Speed']
df.head()
```

```python
# name with highest ratio

df.loc[df['ratio'].max()]['Name']
```

```python
# no of names of ratio > 5
# df.loc[df['ratio']>5].count()
df.loc[df['ratio']>5].shape
```

```python
# max speed of ratio > 5
df.loc[df['ratio']>5].sort_values(by='Speed',ascending=False)
# df.loc[df['ratio']>5,'Speed'].max()
```
